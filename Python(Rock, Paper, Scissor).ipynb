#### Aim:
Write a Python program to develop a Rock, Papers and Scissors game to be played against a computer.

```python
# importing library
import os
import random
import sys

# defining class RPS
class RPS:
    user_choice = ""
    bot_choice = ""
    
#     Defining function clear to clear the output sceen based on operating system 
    def clear(self):
        command = "clear"
        if os.name in ('nt', 'dos'):
            command = "cls"
        os.system(command)

#       Defining menu function 
    def menu(self):
        self.clear()
        print("Welcome to Rock, Paper and Scissors Game")
        print("1. rock")
        print("2. paper")
        print("3. scissors")
        print("4. exit")

#         Defining set_user_choice function to take input from user
    def set_user_choice(self):
        self.user_choice = input("Enter your choice: ")
        if self.user_choice == "exit":
            print("\tGood Bye !!!")
            sys.exit()
        if self.user_choice not in ('rock', 'paper', 'scissors'):
            print("\tInvalid selection !!!")
            input("\tPress any key to continue....")
            self.menu()
            self.set_user_choice()

#             Defining function set_bot_choice for bot selection from Rock, Paper and Scissor
    def set_bot_choice(self):
        self.bot_choice = random.choice(["rock", "paper", "scissors"])

#         Defining function for displaying user and bot choice
    def show_selections(self):
        print("\tUser : " + self.user_choice)
        print("\tBot : " + self.bot_choice)

#         Defining function show_winner 
    def show_winner(self):
        if self.user_choice == self.bot_choice:
            print("\tBoth players selected " + self.user_choice + ". It's a tie!")
        elif self.user_choice == "rock":
            if self.bot_choice == "scissors":
                print("\tRock smashes Scissors! You win!")
            else:
                print("\tPaper covers Rock! You lose!")
        elif self.user_choice == "paper":
            if self.bot_choice == "rock":
                print("\tPaper covers Rock! You win!")
            else:
                print("\tScissors cuts Paper! You lose!")
        elif self.user_choice == "scissors":
            if self.bot_choice == "paper":
                print("\tScissors cuts Paper! You win!")
        else:
            print("\tRock smashes Scissors! You lose!")

def main():
    while(True):
        rps = RPS()
        rps.menu()
        rps.set_user_choice()
        rps.set_bot_choice()
        rps.show_selections()
        rps.show_winner()
        input("\n\tPress any key to continue....")

if __name__ == "__main__":
    main() 
```
